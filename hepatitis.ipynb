### analysis based on steroid 

```python
d = data.groupby(['steroid','class','sex'])['sl_no'].count().reset_index()
```

```python
d
```

```python
label = d['sex']
values = d['sl_no']
```

```python
plt.axis("equal")
colors = ['y','m']
plt.pie(values,labels = label,radius=1.5,colors=colors,autopct='%0.2f%%')
plt.show()
```
